# Collecting data from api

```python
import pandas as pd
from Data.APICalls import get_intervals, get_laps, get_pit_data, get_position, get_session, get_stints, get_weather
import numpy as np
import time
```

## Reading Drivers data

```python
driver_df = pd.read_csv("../Data/drivers.csv")
session_keys = driver_df['session_key']
meeting_keys = driver_df['meeting_key']
driver_numbers = driver_df['driver_number']
```

### Using Driver data to complete api calls

Weather data

```python
meeting_key = 1239
unique_weather_data = set()

for wind_direction in range(250, 330, 10):
    for track_temperature in range(10, 40):
        weather_df = get_weather(meeting_key, wind_direction, track_temperature)

        if not weather_df.empty:
            weather_df_frozen = frozenset(weather_df.itertuples(index=False, name=None))    

            if weather_df_frozen not in unique_weather_data:
                unique_weather_data.add(weather_df_frozen)

unique_weather_dfs = [pd.DataFrame(list(frozen_set)) for frozen_set in unique_weather_data]
combined_weather_df = pd.concat(unique_weather_dfs, ignore_index=True)
combined_weather_df
```

```python
weather_df = get_weather(1239, 250, 10)
weather_df
```

Lap Data

```python
driver_sets = {driver_number: [] for driver_number in driver_numbers}
session_key = 9550

# Collect lap data for each driver
for driver_number in driver_numbers:
    for lap in range(2, 71):
        lap_df = get_laps(session_key, driver_number, lap)

        if not lap_df.empty:
            driver_sets[driver_number].append(lap_df)
        time.sleep(1)

# Convert lists of DataFrames to a single DataFrame for each driver
for driver_number, driver_list in driver_sets.items():
    if driver_list:  # Check if the list is not empty
        driver_sets[driver_number] = pd.concat(driver_list, ignore_index=True)
    else:
        driver_sets[driver_number] = pd.DataFrame()  # If no data, create an empty DataFrame

# Print the resulting DataFrames for each driver
for driver_number, driver_df in driver_sets.items():
    print(f"Driver Number: {driver_number}, DataFrame:\n{driver_df}\n")
```

Interval Data

```python
session_key = 9550
intervals_df = []

for interval in np.arange(0.01, 2.0, 0.01):  
    interval_df = get_intervals(session_key, interval)
    intervals_df.append(interval_df)
    time.sleep(1)

final_intervals_df = pd.concat(intervals_df, ignore_index=True)
final_intervals_df
```

Position Data

```python
meeting_key = 1239
final_position_dfs = []

for driver_num in driver_numbers:
    driver_position_dfs = []  
    for position in range(driver_numbers.size):
        position_df = get_position(meeting_key, driver_num, position)
        driver_position_dfs.append(position_df)
        time.sleep(1)
        
    driver_df_concat = pd.concat(driver_position_dfs, ignore_index=True)
    final_position_dfs.append(driver_df_concat)

final_position_df = pd.concat(final_position_dfs, ignore_index=True)
final_position_df
```

```python
session_key = 9550
stint_df = get_stints(session_key)
stint_df
```

Session Data

```python
session_type = ['Practice 1', 'Practice 2', 'Practice 3', 'Qualifying', 'Race']
year = [2022, 2023, 2024]

all_session_dfs = []

for s_type in session_type:
    session_dfs = []
    for y in year:
        session_df = get_session(s_type, y)
        session_dfs.append(session_df)
        time.sleep(1)
        
    session_df_concat = pd.concat(session_dfs, ignore_index=True)
    all_session_dfs.append(session_df_concat)
    
final_session_df = pd.concat(all_session_dfs, ignore_index=True)
final_session_df
```
